```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
```

```python
df
```

```python
df.set_index("year",inplace=True) # True means the original df changed , no new dataframe created
```

```python
df
```

```python
df.describe()
```

```python
x= df.index
y= df['per capita income (US$)']
plt.xlabel('Year')
plt.ylabel('Per capita Income in $')
plt.scatter(x,y,color='green',marker='.')
```

```python
A = df.index.to_numpy().reshape(-1, 1)  # 2D array for features
B = df['per capita income (US$)']  # 1D array for target
reg= linear_model.LinearRegression()
reg.fit(A,B)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2019]]) # pass a 2 D array as argument
```
