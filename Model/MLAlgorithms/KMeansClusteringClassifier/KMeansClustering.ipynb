```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("income.csv")
df.head()
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_predicted=km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster']=y_predicted
```

```python
df
```

# Need to make the x and y values scalable same i.e. 0 to 1
# Preprocessing using min max scaler

```python
scaler= MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)']=scaler.transform(df[['Income($)']])
df
```

```python

scaler.fit(df[['Age']])
df['Age']=scaler.transform(df[['Age']])
df
```

```python
y_predicted=km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
df1=df[df['cluster']==0]
df2=df[df['cluster']==1]
df3=df[df['cluster']==2]
plt.scatter(df1['Age'],df1['Income($)'], color='Red')
plt.scatter(df2['Age'],df2['Income($)'], color='Green')
plt.scatter(df3['Age'],df3['Income($)'], color='Blue')
```

# Find out the centroids

```python
km.cluster_centers_
```

```python
df1=df[df['cluster']==0]
df2=df[df['cluster']==1]
df3=df[df['cluster']==2]
plt.scatter(df1['Age'],df1['Income($)'], color='Red')
plt.scatter(df2['Age'],df2['Income($)'], color='Green')
plt.scatter(df3['Age'],df3['Income($)'], color='Blue')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], color='Black', marker='*')
plt.legend
```

# - Inertia is a metric used in the k-means clustering algorithm to measure how well a dataset has been clustered. A good model has low inertia and a low number of clusters (K)

```python
k_range= range(1,10,1)
sse =[]
for k in k_range:
    km= KMeans(n_clusters=k)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
sse
              
```

# Elbow Plot

```python
plt.xlabel('K-value')
plt.ylabel('Sum of Squared Errors')
plt.plot(k_range,sse)
```
