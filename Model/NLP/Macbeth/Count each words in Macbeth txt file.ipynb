```python
with open ('macbeth.txt','r') as file:
    d={}
    for line in file: # iterate each line in the file
        line=line.strip()# Remove the leading spaces and newline character
        line=line.lower() # make everything lower case to ease calculation
        words=line.split(' ') # Split line in each words
        for word in words: # iterating each word
            if word not in d: # if a word not found in dictionary
                d[word]=1   # then value assign to 1
            else:   
                d[word]+=1 # else add 1 to the last value of that word found again
    print(d)
```

# Another approach

```python
with open ('macbeth.txt','r') as file:
    d={}
    count=1
    for line in file: # iterate each line in the file
        line=line.strip()# Remove the leading spaces and newline character
        line=line.lower() # make everything lower case to ease calculation
        words=line.split(' ') # Split line in each words
        for word in words: # iterating each word
            if word not in d: # if a word not found in dictionary
                d[word]=count   # then value assign to 1
            else:   
                d[word]+=count # else add 1 to the last value of that word found again
    print(d)
```
