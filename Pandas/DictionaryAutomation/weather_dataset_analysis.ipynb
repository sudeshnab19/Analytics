# Create a list with individual columns and its values as dictionaries.
So if 11 columns and 31 rows are there.
Then 31 iterations needed to show all the 11 columns with values. i.e. 0 to 30

```python
data = []
# dic = {} not defined here as it will only take the last row values not iterate all the values
for r in range(len(df.values)):
    dic={} # dict defined here to iterate all the rows
    for c in range(len(df.columns)):
#         dic={} not defined here as it will only take the last column value not iterate all the columns
        dic[df.columns[c]] = df.values[r,c]  # to show individual cell values for a multi-dimensional array  
    data.append(dic)# append is done here to iterate all the rows here
#data.append(dic) append is not done here as only last row it would have shown
data
```
