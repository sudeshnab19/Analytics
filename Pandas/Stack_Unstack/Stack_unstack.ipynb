# Stack is used for Multi-hierarchy

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('stocks.xlsx')
df
```

# header=[0,1] as 2 headers are there, so we specify 0 and 1
# index_col=0 signifies 1st column to be the index

```python
df=pd.read_excel('stocks.xlsx',header=[0,1],index_col=0)
df
```

```python
df_stack=df.stack()
df_stack
```

```python
df.stack(level=0,dropna=True) # only 1 level headers are there
```

```python
df_stack.unstack() # THe actual excel file
```
